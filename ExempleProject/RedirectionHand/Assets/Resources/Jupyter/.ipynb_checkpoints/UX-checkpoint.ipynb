```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
sns.set(rc={'figure.figsize':(20,8.27)})


nbUsers = 15
startNbUser = 1

allData = pd.DataFrame()
nameList = []

for i in range(startNbUser, startNbUser + nbUsers):
    nameList.append("User"+str(i))
    
for i in range(0, len(nameList)):
    
    results = None
    results_forcedChoice = None
    results_forces = None
    resultsPerUser = None
    
    results = pd.read_csv('../DataCollection/UX/' + nameList[i] + '.csv')
    results['User'] = nameList[i]
    results['Configs'] = results.index    
    results_forcedChoice = pd.read_csv('../DataCollection/UX/' + nameList[i] + '-Confidence.csv')
    
    results_forces = pd.read_csv('../DataCollection/UX/' + nameList[i] + '-MaxForces.csv', sep = ';', decimal = ',')
    pd.to_numeric(results_forces['Maximum Force'])

    
    
    resultsPerUser = pd.concat([results, results_forcedChoice], axis = 1)
    resultsPerUser = pd.concat([resultsPerUser, results_forces], axis = 1)

    resultsPerUser['Confidence'] = resultsPerUser['Results']
    del resultsPerUser['Results']
    allData = pd.concat([resultsPerUser, allData], axis = 0)
    allData.loc[(allData['2-Forced Choice'] == 0) | (allData['2-Forced Choice'] == '0'), 'Similar'] = True
    allData.loc[(allData['2-Forced Choice'] == 1) | (allData['2-Forced Choice'] == '1'), 'Similar'] = False
    
tasks = ['Squeeze', 'Compress']
scales = ['Large', 'Small']
objects = ['Cube']


stiffness = None
range_0 = np.arange(45, 66, 2)
range_1 = np.arange(10, 25, 2)
range_2 = np.arange(1, 10, 5)
range_3 = np.arange(70, 76, 5)
range_4 = np.arange(25, 45, 5)

stiffness = np.append(range_0, range_1)
stiffness = np.append(stiffness, range_2)
stiffness = np.append(stiffness, range_3)
stiffness = np.append(stiffness, range_4)

# stiffness = np.append(np.arange(45, 65, 2), np.arange(10, 25, 2))
# stiffness = np.append(stiffness, np.arange(1, 10, 5))
# stiffness = np.append(stiffness, np.arange(65, 76, 5))
# stiffness = np.append(stiffness, np.arange(25, 45, 5))

for i in range(0, len(results)):
    config = int(results['Configs'].loc[i])
    for l in range(0, len(stiffness)):
        if((config < len(tasks)*len(objects)*len(scales)*(l+1)) &  (config >= len(tasks)*len(objects)*len(scales)*l)):
            stiffnessToTouch = stiffness[l]
            for k in range(0, len(scales)):
                if((config < len(tasks)*len(objects)*len(scales)*l + (k+1)*len(tasks)*len(objects)) & (config >= len(tasks)*len(objects)*len(scales)*l + k*len(tasks)*len(objects))):
                    scaleObj = scales[k]
                    for j in range(0, len(objects)):
                        if((config < (l*len(tasks)*len(objects)*len(scales) + k*len(tasks)*len(objects) + (j+1)*len(tasks))) & (config >= (l*len(tasks)*len(objects)*len(scales) + k*len(tasks)*len(objects) + j*len(tasks)))):
                            objToCatch = objects[j]
                            for m in range(0, len(tasks)):
                                if((config < (l*len(tasks)*len(objects)*len(scales) + k*len(tasks)*len(objects) + j*len(tasks) + m+1)) & (config >= (l*len(tasks)*len(objects)*len(scales) + k*len(tasks)*len(objects) + j*len(tasks) + m))):
                                    taskToDo = tasks[m]
                  
    allData.loc[allData['Configs'] == config, 'Task'] = taskToDo
    allData.loc[allData['Configs'] == config, 'Scale'] = scaleObj
    allData.loc[allData['Configs'] == config, 'Stiffness'] = stiffnessToTouch
    pd.to_numeric(allData['Confidence'])
    pd.to_numeric(allData['2-Forced Choice'])

    allData.loc[allData['2-Forced Choice'] == 0, '2AFC-Sim1'] = 1
    allData.loc[allData['2-Forced Choice'] == 1, '2AFC-Sim1'] = 0
    
    allData.loc[(allData['Task'] == 'Squeeze') & (allData['Scale'] == 'Large'), 'Scales and Task'] = 'Large Object, Squeeze'
    allData.loc[(allData['Task'] == 'Squeeze') & (allData['Scale'] == 'Small'), 'Scales and Task'] = 'Small Object, Squeeze'
    allData.loc[(allData['Task'] == 'Compress') & (allData['Scale'] == 'Large'), 'Scales and Task'] = 'Large Object, Compress'
    allData.loc[(allData['Task'] == 'Compress') & (allData['Scale'] == 'Small'), 'Scales and Task'] = 'Small Object, Compress'

    
```

```python
stackMe = pd.DataFrame()
for i in range(0, len(stiffness)):
    total = 0
    total_sim = 0
    stackMe.loc[i, 'Stiffness'] = stiffness[i]

    for j in range(1, 6):
        name = 'Sim-Count-' + str(j)
        stackMe.loc[i, name] = len(allData[(allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 1)])
    
        name = 'Diff-Count-' + str(j)
        stackMe.loc[i, name] = len(allData[(allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 0)])
        
        total = total + len(allData[(allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 0)]) + len(allData[(allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 1)])
        total_sim = total_sim + len(allData[(allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 1)])

    stackMe.loc[i, 'TotalSim'] = total_sim
    stackMe.loc[i, 'Total'] = total
    for k in range(0, 2):
        for l in range(0, 2):
            total = 0
            total_sim = 0
            for j in range(1, 6):
                condSize = (allData['Scale'] == scales[l])

                name = 'Sim-Count-' + str(j) + '-' + scales[l]
                cond1 = (allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 1)
                stackMe.loc[i, name] = len(allData[cond1 & condSize])

                name = 'Diff-Count-' + str(j) + '-' + scales[l]
                cond0 = (allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 0)
                stackMe.loc[i, name] = len(allData[cond0 & condSize])

                total = total + len(allData[cond1 & condSize]) + len(allData[cond0 & condSize])
                total_sim = total_sim + len(allData[cond1 & condSize])

            stackMe.loc[i, 'TotalSim-'+scales[l]] = total_sim
            stackMe.loc[i, 'Total-'+scales[l]] = total
            total = 0
            total_sim = 0
            for j in range(1,6):
                condTask = (allData['Task'] == tasks[k])

                name = 'Sim-Count-' + str(j) + '-' + tasks[k]
                cond1 = (allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 1)
                stackMe.loc[i, name] = len(allData[cond1 & condTask])

                name = 'Diff-Count-' + str(j) + '-' + tasks[k]
                cond0 = (allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 0)
                stackMe.loc[i, name] = len(allData[cond0 & condTask])

                total = total + len(allData[cond1 & condTask]) + len(allData[cond0 & condTask])
                total_sim = total_sim + len(allData[cond1 & condTask])

            stackMe.loc[i, 'TotalSim-'+tasks[k]] = total_sim
            stackMe.loc[i, 'Total-'+tasks[k]] = total
            total = 0
            total_sim = 0
            for j in range(1,6):
                
                condTask = (allData['Task'] == tasks[k])
                condSize = (allData['Scale'] == scales[l])

                name = 'Sim-Count-' + str(j) + '-' + tasks[k] + '-' + scales[l]
                cond1 = (allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 1)
                stackMe.loc[i, name] = len(allData[cond1 & condTask & condSize])

                name = 'Diff-Count-' + str(j) + '-' + tasks[k] + '-' + scales[l]
                cond0 = (allData['Confidence'] == j) & (allData['Stiffness'] == stiffness[i]) & (allData['2AFC-Sim1'] == 0)
                stackMe.loc[i, name] = len(allData[cond0 & condTask & condSize])

                total = total + len(allData[cond1 & condTask & condSize]) + len(allData[cond0 & condTask & condSize])
                total_sim = total_sim + len(allData[cond1 & condTask & condSize])

            stackMe.loc[i, 'TotalSim-'+tasks[k]+'-'+scales[l]] = total_sim
            stackMe.loc[i, 'Total-'+tasks[k]+'-'+scales[l]] = total

stackMe['doubt-Squeeze'] = stackMe['Diff-Count-1-Squeeze'] + stackMe['Sim-Count-1-Squeeze']
stackMe['doubt-Compress'] = stackMe['Diff-Count-1-Compress'] + stackMe['Sim-Count-1-Compress']
stackMe['doubt-Small'] = stackMe['Diff-Count-1-Small'] + stackMe['Sim-Count-1-Small']
stackMe['doubt-Large'] = stackMe['Diff-Count-1-Large'] + stackMe['Sim-Count-1-Large']
stackMe['doubt-Squeeze-Small'] = stackMe['Diff-Count-1-Squeeze-Small'] + stackMe['Sim-Count-1-Squeeze-Small']
stackMe['doubt-Squeeze-Large'] = stackMe['Diff-Count-1-Squeeze-Large'] + stackMe['Sim-Count-1-Squeeze-Large']
stackMe['doubt-Compress-Small'] = stackMe['Diff-Count-1-Compress-Small'] + stackMe['Sim-Count-1-Compress-Small']
stackMe['doubt-Compress-Large'] = stackMe['Diff-Count-1-Compress-Large'] + stackMe['Sim-Count-1-Compress-Large']
    
    
    
    
    
stackMe = stackMe.sort_values(by = 'Stiffness')
stackMe = stackMe.reset_index()
del stackMe['index']
stackMe['doubt'] = stackMe['Diff-Count-1'] + stackMe['Sim-Count-1']
stackMe
stackMe.head()
```

```python
sortForces = pd.DataFrame()
zvalue = 1.96

sortForces['All'] = allData.groupby('Stiffness')['Maximum Force'].mean()
sortForces['All_std'] = allData.groupby('Stiffness')['Maximum Force'].std()
sortForces['All_count'] = allData.groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'All-95CI'] = zvalue*sortForces['All_std'].values[i]/math.sqrt(sortForces['All_count'].values[i])


cond0 = (allData['Task'] == 'Squeeze')
sortForces['Squeeze'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Squeeze_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Squeeze_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Squeeze-95CI'] = zvalue*sortForces['Squeeze_std'].values[i]/math.sqrt(sortForces['Squeeze_count'].values[i])

cond0 = (allData['Task'] == 'Compress')
sortForces['Compress'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Compress_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Compress_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Compress-95CI'] = zvalue*sortForces['Compress_std'].values[i]/math.sqrt(sortForces['Compress_count'].values[i])

cond0 = (allData['Task'] == 'Squeeze') & (allData['Scale'] == 'Small')
sortForces['Squeeze-Small'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Squeeze-Small_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Squeeze-Small_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Squeeze-Small-95CI'] = zvalue*sortForces['Squeeze-Small_std'].values[i]/math.sqrt(sortForces['Squeeze-Small_count'].values[i])

cond0 = (allData['Task'] == 'Compress') & (allData['Scale'] == 'Small')
sortForces['Compress-Small'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Compress-Small_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Compress-Small_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Compress-Small-95CI'] = zvalue*sortForces['Compress-Small_std'].values[i]/math.sqrt(sortForces['Compress-Small_count'].values[i])

cond0 = (allData['Task'] == 'Squeeze') & (allData['Scale'] == 'Large')
sortForces['Squeeze-Large'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Squeeze-Large_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Squeeze-Large_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Squeeze-Large-95CI'] = zvalue*sortForces['Squeeze-Large_std'].values[i]/math.sqrt(sortForces['Squeeze-Large_count'].values[i])

cond0 = (allData['Task'] == 'Compress') & (allData['Scale'] == 'Large')
sortForces['Compress-Large'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Compress-Large_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Compress-Large_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Compress-Large-95CI'] = zvalue*sortForces['Compress-Large_std'].values[i]/math.sqrt(sortForces['Compress-Large_count'].values[i])

cond0 = (allData['Scale'] == 'Small')
sortForces['Small'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Small_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Small_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Small-95CI'] = zvalue*sortForces['Small_std'].values[i]/math.sqrt(sortForces['Small_count'].values[i])

cond0 =(allData['Scale'] == 'Large')
sortForces['Large'] = allData[cond0].groupby('Stiffness')['Maximum Force'].mean()
sortForces['Large_std'] = allData[cond0].groupby('Stiffness')['Maximum Force'].std()
sortForces['Large_count'] = allData[cond0].groupby('Stiffness')['Maximum Force'].count()
for i in range(0, len(sortForces)):
    sortForces.loc[stiffness[i], 'Large-95CI'] = zvalue*sortForces['Large_std'].values[i]/math.sqrt(sortForces['Large_count'].values[i])
```
